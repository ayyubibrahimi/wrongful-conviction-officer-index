```python
import pandas as pd
```

```python
def read_allegations(): 
    df = pd.read_csv("../data/input/allegation.csv")
    return df 

dfa = read_allegations()
dfa = dfa.rename(columns={"uid": "llead_uid"})

def read_wcoi():
    df = pd.read_csv("../../wcoi-post/data/output/wcoi_post.csv")
    return df 

dfb = read_wcoi()

dfb = dfb.rename(columns={"llead_uid1": "llead_uid"})

dfb = dfb[["first_name2", "last_name2", "agency1", "llead_uid", "filename2", "page_number2"]]

dfb = dfb[dfb.agency1.str.contains("orleans")]

df = pd.merge(dfa, dfb, on="llead_uid")

df = df.drop_duplicates(subset=["allegation_uid"])

df.llead_uid.value_counts(ascending=False).head(30)

df.llead_uid.nunique()
```

```python
df.to_csv("../data/output/wcoi-llead-post-allegation-table.csv", index=False)
```

```python
# Group by the relevant columns and count llead_uid occurrences
grouped_df = df.groupby(['first_name2', 'last_name2', 'filename2', 'page_number2']).size().reset_index(name='llead_uid_count')

# Sort by llead_uid_count in descending order
sorted_df = grouped_df.sort_values(by='llead_uid_count', ascending=False)

sorted_df
```
